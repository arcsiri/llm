```python
#FAISS

from langchain_community.document_loaders import PyPDFLoader
from langchain_community.vectorstores import FAISS
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import OllamaEmbeddings

loader = PyPDFLoader("my.pdf")
docs=loader.load()
docs

text_splitter=RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=50)
final_docs=text_splitter.split_documents(docs)
final_docs
```

```python
embedding=OllamaEmbeddings(model="gemma:2b")
db=FAISS.from_documents(final_docs,embedding)
```

```python
query="Learn the benefits, typical use cases, and technical concepts of Amazon Keyspaces. You can try the service through the sample code provided or the interactive tool in the AWS Management Console."
docs=db.similarity_search(query)
docs
```

```python
docs[0]
```

```python
docs[0].page_content
```

```python
#as the retreiver

retreiver=db.as_retriever()
retreiver.invoke(query)
```

```python
#specality of retreiver
```

```python
#similarity score
docs_and_score=db.similarity_search_with_score(query)
docs_and_score
```

```python
#it return the distance ie manhatten score in the form of score
#so lower the score the better similarity
```

```python
#in the above it would compare the documents
```

```python
# but we can also compare the vectors
```

```python
embedding_vectors=embedding.embed_query(query)
```

```python
embedding_vectors
```

```python
docs_score=db.similarity_search_by_vector(embedding_vectors)
docs_score
```

```python
#save the vercor db in local
#loading into local
db.save_local("faiss_index")#it will just create a folder with the name
```

```python
new_db=FAISS.load_local("faiss_index",embedding,allow_dangerous_deserialization=True)
```

```python
docs=new_db.similarity_search(query)
```

```python
from langchain_chroma import Chroma

from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import OllamaEmbeddings

loader = PyPDFLoader("my.pdf")
docs=loader.load()
docs

text_splitter=RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=50)
final_docs=text_splitter.split_documents(docs)
final_docs

embedding=OllamaEmbeddings(model="gemma:2b")
db=Chroma.from_documents(final_docs,embedding)

```

```python
query="Learn the benefits, typical use cases, and technical concepts of Amazon Keyspaces. You can try the service through the sample code provided or the interactive tool in the AWS Management Console."
docs=db.similarity_search(query)
docs
```

```python
#saving to the local
vectordb=Chroma.from_documents(documents=final_docs,embedding=embedding,persist_directory="./chroma_db")
```

```python
#load to disk
db2=Chroma(persist_directory=".\chroma_db",embedding_function=embedding)
docs=db.similarity_search(query)
docs
```

```python
print(docs[0].page_content)
#sismilarly we cal also compare the vectore
```
